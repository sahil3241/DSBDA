## Data Preprocessing

###### Dropping rows having math score value as '?'

```python
df.drop(df.index[df['math score'] == '?'], inplace=True)
```

```python
df = df.astype({'math score' : float})
```

```python
df.info()
```

```python
df['math score'].fillna(df['math score'].mean(), inplace=True)
```

```python
df['reading score'].fillna(df['reading score'].mean(), inplace=True)
```

```python
df['writing score'].fillna(df['writing score'].mean(), inplace=True)
```

```python
df.isnull().sum()
```

## Outliers

```python
# IQR Method of Removing Outliers

plt.figure(figsize=(6, 8))
sns.boxplot(y=df['math score'])
plt.ylabel('math_score_float')
plt.title('Boxplot of math')
plt.show()

# Calculate IQR
Q1 = df['math score'].quantile(0.25)
Q3 = df['math score'].quantile(0.75)
IQR = Q3 - Q1

# Define limits to identify outliers
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

# Remove outliers
df = df[(df['math score'] >= lower_limit) & (df['math score'] <= upper_limit)]

# Create a boxplot again without outliers
plt.figure(figsize=(6, 8))
sns.boxplot(y=df['math score'])
plt.ylabel('math score')
plt.title('Boxplot of math score without outliers')
plt.show()
```

```python
plt.figure(figsize=(6, 8))
sns.boxplot(y=df['reading score'])
plt.ylabel('reading score')
plt.title('Boxplot of reading score')
plt.show()

# Calculate IQR
Q1 = df['reading score'].quantile(0.25)
Q3 = df['reading score'].quantile(0.75)
IQR = Q3 - Q1

# Define limits to identify outliers
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

# Remove outliers
df = df[(df['reading score'] >= lower_limit) & (df['reading score'] <= upper_limit)]

# Create a boxplot again without outliers
plt.figure(figsize=(6, 8))
sns.boxplot(y=df['reading score'])
plt.ylabel('reading score')
plt.title('Boxplot of SalePrice without outliers')
plt.show()
```

```python
plt.figure(figsize=(6, 8))
sns.boxplot(y=df['writing score'])
plt.ylabel('writing score')
plt.title('Boxplot of writing score')
plt.show()

# Calculate IQR
Q1 = df['writing score'].quantile(0.25)
Q3 = df['writing score'].quantile(0.75)
IQR = Q3 - Q1

# Define limits to identify outliers
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

# Remove outliers
df = df[(df['writing score'] >= lower_limit) & (df['writing score'] <= upper_limit)]

# Create a boxplot again without outliers
plt.figure(figsize=(6, 8))
sns.boxplot(y=df['writing score'])
plt.ylabel('writing score')
plt.title('Boxplot of SalePrice without outliers')
plt.show()
```
