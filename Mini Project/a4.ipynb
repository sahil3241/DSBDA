```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
covid1 =pd.read_csv('covid_vaccine_statewise.csv')
```

```python
covid2 = covid1[covid1['State'] != 'India']
# print()
covid2.head(len(covid2))
```

```python
covid2.describe()
```

```python
state_groupby = covid2.groupby(by='State')['First Dose Administered'].sum()
state_groupby.head(100)
```

```python
plt.figure(figsize=(8, 6))
state_groupby.plot(kind='bar')
plt.title('Total Number of First Doses Administered by State')
plt.xlabel('State')
plt.ylabel('Total Number of First Doses')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
state_groupby1 = covid2.groupby(by='State')['Second Dose Administered'].std()
state_groupby1.head(100)
```

```python
plt.figure(figsize=(8, 6))
state_groupby1.plot(kind='bar', color='green')  # Using a different color for second dose
plt.title('Total Number of Second Doses Administered by State')
plt.xlabel('State')
plt.ylabel('Total Number of Second Doses')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
specific_date_str = '09/08/2021'

specific_date_entries = covid2[covid2['Updated On'] == specific_date_str].copy()

total_male_doses_on_specific_date = specific_date_entries['Male (Doses Administered)'].sum()

print(total_male_doses_on_specific_date)
```

# Number of persons state wise vaccinated for first dose in India

```python
state_groupby_male_doses = specific_date_entries.groupby('State')['Male (Doses Administered)'].sum()
plt.figure(figsize=(8, 6))
state_groupby_male_doses.plot(kind='bar', color='red')
plt.title('Total Male Doses Administered by State')
plt.xlabel('State')
plt.ylabel('Total Male Doses')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
specific_date_str = '09/08/2021'

specific_date_entries = covid2[covid2['Updated On'] == specific_date_str].copy()

total_female_doses_on_specific_date = specific_date_entries['Female (Doses Administered)'].sum()

print(total_female_doses_on_specific_date)
```

```python
state_groupby_female_doses = specific_date_entries.groupby('State')['Female (Doses Administered)'].sum()
plt.figure(figsize=(8, 6))
state_groupby_female_doses.plot(kind='bar', color='black')
plt.title('Total Female Doses Administered by State')
plt.xlabel('State')
plt.ylabel('Total Female Doses')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

# Number of deaths , cured , and Confirmed


```python
covid1 =pd.read_csv('covid_vaccine_statewise.csv')
covid_deaths = pd.read_csv('covid_19_india.csv')
covid_test = pd.read_csv('StatewiseTestingDetails.csv')
```

```python
covid_deaths.head(len(covid_deaths))
```

```python
D_stateGroupby = covid_deaths.groupby(by='State/UnionTerritory')['Deaths'].sum()
D_stateGroupby.head(100)

```

```python
plt.figure(figsize=(8, 6))
D_stateGroupby.plot(kind='bar')
plt.title('TDeaths')
plt.xlabel('State')
plt.ylabel('Deaths')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
confirm_stateGroupby = covid_deaths.groupby(by='State/UnionTerritory')['Confirmed'].sum()
confirm_stateGroupby.head(100)
```

```python
plt.figure(figsize=(8, 6))
confirm_stateGroupby.plot(kind='bar')
plt.title('Confirmed')
plt.xlabel('State')
plt.ylabel('Confirmed Cases')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
cured_stateGroupby = covid_deaths.groupby(by='State/UnionTerritory')['Cured'].sum()
cured_stateGroupby.head(100)
```

```python
plt.figure(figsize=(8, 6))
cured_stateGroupby.plot(kind='bar')
plt.title('Cured')
plt.xlabel('State')
plt.ylabel('Cured')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
